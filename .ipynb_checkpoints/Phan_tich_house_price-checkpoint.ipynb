# 2. Lọc ra các bản ghi bán nhà riêng tại phường Trung Liệt hoặc phường Khâm Thiên

```python
df["type_of_land"].unique()
```

```python
df["type_of_land"].replace('Bán nhà riêng\n ', 'Bán nhà riêng', inplace = True)
df["type_of_land"].replace('Bán nhà mặt phố\n ', 'Bán nhà mặt phố', inplace = True)
df["type_of_land"].replace('Bất động sản khác\n ', 'Bất động sản khác', inplace = True)
df["type_of_land"].unique()
```

```python
df_nharieng = df[df.type_of_land == 'Bán nhà riêng']
df_nharieng.head()
```

```python
df["ward_name"].unique()
```

```python
df_nharieng.info()
```

```python
final_df = pd.concat([df_nharieng[df.ward_name == 'Phường Trung Liệt'], df_nharieng[df.ward_name == 'Phường Khâm Thiên']], axis = 0)
final_df.head()
```

# 4. Với mỗi loại nhà đất, tính trung bình cộng giá cũng như giá lớn nhất và giá nhỏ nhất

```python
df.info()
```

```python
df["type_of_land"].unique()
```

```python

df.groupby("type_of_land")["price"].mean()
```

```python
print("Giá nhỏ nhất: ")
df.groupby("type_of_land")["price"].min()
```

```python
print("Giá lớn nhất: ")
df.groupby("type_of_land")["price"].max()
```
